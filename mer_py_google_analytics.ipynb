```python
import matplotlib.pyplot as plt
import numpy as np
%pylab inline
```

```python
def is_header(line):
    return '#' == line[0]
```

```python
def is_data(line):
    return line.startswith('/')
#    is_titles = line.startswith('Page')
#    is_data = not is_header(line) and (not is_titles) and ''!=line.strip()
#    return is_data
```

```python
def page_type(line):
    if 3 == line.count('/'):
        return 'exam_page'
    if 4 == line.count('/'):
        return('course_page')
    if 5 == line.count('/'):
        return('question_page')
    print line
    return 'other'
```

```python
def average_time_on_page(line):
    try:
        avg_time_line = line.split('\t')[4]
        hours,minutes,seconds = avg_time_line.split(':')
    except ValueError:
        print line
        return 0

    return float(seconds) + 60*float(minutes) + 60*60*float(hours)
    
```

```python
def plot_pie_dict(my_dict,my_title):
    sorted_keys = sort(my_dict.keys())
    y = np.zeros(len(sorted_keys))
    my_labels = []
    for num,key in enumerate(sorted_keys):
        y[num] = my_dict[key]
        my_labels.append(key + ' [' + str(int(my_dict[key])) + 's]')
    plt.pie(y,labels = my_labels,startangle=90)
    plt.title(my_title)
```

```python
def add_page_to_all_exams(all_exams,line):
    exam_name = get_exam_name(line)
    question_name = get_question_name(line)
    if not exam_name in all_exams.keys():
        all_exams[exam_name] = {}
    current_exam_dict = add_to_current_exam_dict(all_exams[exam_name],question_name,line)
    all_exams[exam_name] = current_exam_dict
    return all_exams
```

```python
def add_to_current_exam_dict(current_dict,question_name,line):
    average_time = average_time_on_page(line)
    current_dict[question_name] = average_time
    return current_dict
```

```python
def get_question_name(line):
    question_name = line.split('\t')[0].split('/')[-1].replace('_',' ').replace('Question ','Q')
    return question_name
```

```python
def get_exam_name(line):
    exam_name = line.split('\t')[0].replace('/Science:Math_Exam_Resources/Courses/MATH','')
    course,term = exam_name.split('/')[0:2]
    exam_name = 'Math ' + course + ', ' + term.replace('_',' ')
    return exam_name
```

```python
average_time = {'exam_page': 0, 'course_page': 0, 'question_page': 0}
all_exams = {}
f = open('merdata.csv','r')


for num,line in enumerate(f):
    if not is_data(line):
        continue
    average_time[page_type(line)] += average_time_on_page(line)
    if 'question_page' == page_type(line):
        all_exams = add_page_to_all_exams(all_exams,line)    
    
plot_pie_dict(average_time,'Average time on page')

```

```python
for num,key in enumerate(all_exams.keys()):
    fig, ax = plt.subplots(1,1)
    plot_pie_dict(all_exams[key],key)
    fig.savefig(key + '.png')
```
